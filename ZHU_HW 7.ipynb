Simple vs. Multiple Linear Regression:

    Simple Linear Regression (SLR) models the relationship between one independent variable (predictor) and one dependent variable (outcome). It assumes a straight-line (linear) relationship. On the other hand, Multiple Linear Regression (MLR) models the relationship between multiple independent variables and one dependent variable. It presents more complex relationships and  improves predictive power by accounting for potential confounding factors.
    
Continuous vs. Indicator Variable in SLR:

    A continuous variable is based on continuous data, data that can take any numerical value in a range (e.g., height, age). In SLR, it's used as a predictor to model a continuous relationship with the outcome. An indicator variable is based on categorical data. It takes binary values (0 or 1) to represent different groups or categories. When used in SLR, it models the difference in means between two categories, assuming the relationship with the outcome is linear.

Effect of Adding an Indicator Variable in MLR:

    When an indicator variable is added alongside a continuous variable in MLR, it introduces the effect of the categorical group on the outcome, while still modeling the relationship with the continuous variable. The model becomes more nuanced, showing how both types of predictors influence the dependent variable. In SLR, the effect of a categorical variable is modeled as a simple shift (e.g., comparing two groups). In MLR, the indicator variable models this shift while also accounting for the continuous variable's effect.

Interaction Between Continuous and Indicator Variables in MLR:

    Adding an interaction term between a continuous and an indicator variable in MLR models how the effect of the continuous variable on the outcome changes depending on the category of the indicator variable. The model captures the different slopes for the continuous variable in each category.

Indicator Variables from Non-Binary Categorical Data in MLR:

    When using a non-binary categorical variable (e.g., three or more categories) in MLR, it‚Äôs typically encoded into binary indicator variables. These indicator variables allow the model to estimate different intercepts for each category of the non-binary variable, essentially comparing the mean of the dependent variable across different categories. For example, a categorical variable Color with three categories (Red, Blue, Green) is encoded into two binary indicator variables, representing Red and Blue, with Green as the reference category: ùëå=ùõΩ0+ùõΩ1‚ãÖRed+ùõΩ2‚ãÖBlue+ùúñ. Variable Red will be 1 when observing the Red category and zero otherwise. The same for the varible Blue when observing the Blue category. The coefficient ùõΩ0 represents the intercept, or the expected value of ùëå when Color is Green (the reference category). ùõΩ1 represents the change in ùëå when Color is Red compared to Green. ùõΩ2 represents the change in ùëå when Color is Blue compared to Green.

Formulas:
- If there is no interaction effect, then the model would assume that the effects of TV and online advertising budgets are independent of each other. The equation is: Y = Œ≤0 + Œ≤1‚ãÖX1 + Œ≤2‚ãÖX2 + œµ.

  - Œ≤0: Intercept (baseline effectiveness when spend for both TV and online ads are at zero).
  - Œ≤1: Coefficient for TV advertising spend (the change in effectiveness for each additional unit of TV ad spend).
  - Œ≤2: Coefficient for online advertising spend (the change in effectiveness for each additional unit of online ad spend).
  - Œµ: Error term.
  - Œ≤3: Coefficient for the interaction term, representing how the effect of one predictor (TV ads) on the outcome changes depending on the level of the other predictor (online ads).

Predictions using each model:

- Without Interaction: Predictions from the model assume that the effects of TV and online advertising are independent of each other. The model predicts the outcome by considering how the combination of TV and online ad spend influences the outcome. For example, the effect of increasing TV ad spend may be stronger if online ad spend is also high, or it might be weaker if online ad spend is low.

The Design Matrix is a table where rows correspond to observations (pokeman data in pokeaman_train dataset) and columns correspond to the predictor variables (Attack, Defense, Speed, Legendary, and their interactions).
For interaction terms, like Attack * Defense * Speed * Legendary, the design matrix will include columns for all combinations of the interactions between these predictors, such as Attack * Defense, Attack * Speed, and so on. With many predictors, this results in a large number of columns.

Multicollinearity occurs when predictors in the design matrix, such as Attack, Defense, and Speed, are highly correlated, leading to redundant information and difficulty in determining the most important variables. This results in inflated standard errors for the regression coefficients, making it hard to assess the true effects of each predictor. The model may fit well to the training data but poorly to unseen data, as seen in the low out-of-sample R-squared value, hindering its ability to generalize.

For a design matrix, the condition number found in .summary() measures multicollinearity, with a large value indicating numerical instability. In model3_linear_form, the model without interactions, the condition number was 343. This suggests moderate multicollinearity. After centering and scaling continuous variables like Attack and Defense, the condition number improved to 1.66, indicating reduced multicollinearity. However, for model4_linear_form, the model with interactions, the condition number was 12 trillion. After centering and scaling, the condition number went down to the still extremely large value of 2.25 trillion. This suggested that there is significant instability due to complex interactions between predictors. This multicollinearity causes overfitting, where the model fits well to training data but performs poorly on unseen data. It also inflates the variance of regression coefficients, making predictions highly sensitive to small data changes and impairing the model's generalization ability.

````python
# Code from ChatGPT which was heavily inspired by previous code in this HW
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from patsy import center, scale


# Load data
url = "https://raw.githubusercontent.com/KeithGalli/pandas/master/pokemon_data.csv"
pokeaman = pd.read_csv(url)
pokeaman.fillna('None', inplace=True)
fifty_fifty_split_size = int(pokeaman.shape[0]*0.5)

# Define the model formula
# ChatGPT didn't do this but I am borrowing model3
model3_linear_form_CS = \
  'HP ~ scale(center(Attack)) + scale(center(Defense))' 
# "Cond. No." is NOW 1.66 due to centering and scaling

# Number of iterations
iterations = 100
in_sample_r2_scores = []
out_of_sample_r2_scores = []

# Run the model performance loop
for _ in range(iterations):
    pokeaman_train,pokeaman_test = \
      train_test_split(pokeaman, train_size=fifty_fifty_split_size)

    # Fit the model on training data
    model = smf.ols(formula=model3_linear_form_CS, data=pokeaman_train)
    model_fit = model.fit()

    # Calculate in-sample R-squared
    in_sample_r2 = model_fit.rsquared
    in_sample_r2_scores.append(in_sample_r2)

    # Calculate out-of-sample R-squared on test data
    yhat_test = model_fit.predict(pokeaman_test)
    y_test = pokeaman_test.HP
    out_of_sample_r2 = np.corrcoef(y_test, yhat_test)[0, 1]**2
    out_of_sample_r2_scores.append(out_of_sample_r2)

# Visualize results
plt.figure(figsize=(12, 6))
plt.plot(in_sample_r2_scores, label='In Sample R-squared', marker='o', linestyle='-', markersize=4)
plt.plot(out_of_sample_r2_scores, label='Out of Sample R-squared', marker='x', linestyle='--', markersize=4)
plt.xlabel('Iteration')
plt.ylabel('R-squared')
plt.title('Model Performance Across Multiple Random Splits')
plt.legend()
plt.show()

diff_r2 = []
for i in range(len(in_sample_r2_scores)):
    diff = abs(in_sample_r2_scores[i] - out_of_sample_r2_scores[i])
    diff_r2.append(diff)
print("Average Difference in R-squared:  ", np.mean(diff_r2))
````

However, with R-squared values mostly below 0.3, the model captures only a small portion of the variance, indicating limited predictive power for this dataset.

2. Low Predictive Power: Both R-squared values are low (around 0.15-0.25), indicating the model only explains a limited portion of the variance in the target variable. This could either be because the predictors are weak or because there are other important factors not included. We already established earlier in the HW that it is the latter so we can attribute the low R-squared values to a limited amount of predictors. 

3. High Variability Across Splits: R-squared values fluctuate significantly across iterations, showing the model's sensitivity to different data splits, which might imply instability or variability in the data.

Purpose of Demonstration:
- Without fixing a random seed, each iteration provides a new random split of training and test data. The purpose of this demonstration is to emphasize model robustness across different data splits.
