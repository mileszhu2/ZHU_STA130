```python
# For question 4 I will load a new dataset with more missing values
import pandas as pd
link="https://raw.githubusercontent.com/mwaskom/seaborn-data/master/titanic.csv" 
mn=pd.read_csv(link)
mn
```

```python
# Finding the missing values
mn.isna().sum()
```

```python
# Step 1: Delete the 'deck' column since it has too many missing values
del mn['deck']


# Step 2: Drop rows that contain any missing values in remaining columns
mn_clean = mn.dropna()

# Report the result
print(f"Before: {mn.shape}")  
print(f"After: {mn_clean.shape}") 
```

Before: There are 819 rows and 15 columns. Column 'age' has around 20% missing values while 'deck' has around 77% missing values. There are also rows missing values in the 'embarked' and 'embarked_town columns'.

After: The dataset now has 1 less column and 179 fewer rows. This is because I removed one column 'deck' and removed 177 rows with no value in 'age' and 2 rows with no value in 'embarked' and 'embarked_town'. 

Justification: Like mentioned in the previous question, removing insignificant columns first will preserve the number of people recorded in this dataset.
