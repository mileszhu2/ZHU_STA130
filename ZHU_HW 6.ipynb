The theoretical Simple Linear Regression (SLR) model describes the relationship between two variables: a predictor (or independent) variable, often denoted as x, and an outcome (or dependent) variable, y.

Components of Simple Linear Regression:
Predictor Variable (x): This is the variable we use to predict y. For example, if we are predicting stress based on workload, then workload is x.

Outcome Variable (y): This is the variable we want to predict. For example, if we are predicting stress based on workload, then stress is y.

Slope (𝛽1): This coefficient represents the change in the outcome variable y for each one-unit increase in x. The slope tells us the direction and strength of the relationship between x and y.

Intercept (𝛽0): This is the value of y when x is zero. It's the point where the regression line crosses the y-axis.

Error Term (ϵ): The error term accounts for the random variability in y that the predictor variable x doesn’t explain. In SLR, we assume that this error term is normally distributed with a mean of zero and some standard deviation.

The SLR equation is y = 𝛽0 + 𝛽1(x) + ϵ. This equation states that the outcome variable for a given observation y is equal to the intercept plus the slope times the predictor variable x for that observation, plus a random error term. The error term is assumed to come from a normal distribution with mean 0 and standard deviation σ. In this model, each value of x and y is calculated by adding a random error term ϵ from a normal distribution. This produces values of y that are samples of a normal distribution centered around 𝛽0 + 𝛽1(x), where the spread is controlled by σ.

````python
# Code slightly modified
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

# Set the random seed for reproducibility
np.random.seed(0)

# Define model parameters
n = 100  # Sample size
beta0 = 5  # Intercept
beta1 = 2  # Slope
sigma = 1.5  # Standard deviation of errors

# Generate predictor variable x (e.g., 100 points evenly spaced between 0 and 10)
x = np.linspace(0, 10, n)

# Generate normally distributed error term epsilon
epsilon = norm.rvs(loc=0, scale=sigma, size=n)

# Calculate the outcome variable y
y = beta0 + beta1 * x + epsilon

# Plot the data points
plt.scatter(x, y, color='blue', label='Data points (y)')
plt.plot(x, beta0 + beta1 * x, color='red', label='True regression line (without error)')
plt.xlabel('Predictor Variable (x)')
plt.ylabel('Outcome Variable (y)')
plt.title('Simple Linear Regression Simulation')
plt.legend()
plt.show()
````

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.formula.api as smf

# Set the random seed for reproducibility
np.random.seed(0)

# Define model parameters
n = 100  # Sample size
beta0 = 5  # Intercept
beta1 = 2  # Slope
sigma = 1.5  # Standard deviation of errors

# Generate predictor variable x (e.g., 100 points evenly spaced between 0 and 10)
x = np.linspace(0, 10, n)

# Generate normally distributed error term epsilon
epsilon = norm.rvs(loc=0, scale=sigma, size=n)

# Calculate the outcome variable y
y = beta0 + beta1 * x + epsilon

# Step 1: Create a pandas DataFrame from the simulated data
data = pd.DataFrame({'x': x, 'y': y})

# Step 2: Fit a Simple Linear Regression model using statsmodels
# The 'y ~ x' specifies y as the outcome and x as the predictor value
# The ols function fits an Ordinary Least Squares (OLS) regression, which is the basis of Simple Linear Regression
# The fit() method performs the fitting and returns a model object containing the results
model = smf.ols('y ~ x', data=data).fit()

# Step 3: Print the model summary to see the regression coefficients and fit statistics
# The model.summary() provides a detailed output
# This includes estimated coefficients (intercept and slope), standard errors, p-values, R-squared, and other
print(model.summary())

# Step 4: Visualize the data points and fitted regression line
plt.figure(figsize=(10, 6))
plt.scatter(data['x'], data['y'], color='blue', label='Data points')
plt.plot(data['x'], model.fittedvalues, color='red', label='Fitted regression line')
plt.xlabel('Predictor Variable (x)')
plt.ylabel('Outcome Variable (y)')
plt.title('Fitted Simple Linear Regression Model')
plt.legend()
plt.show()
````

The .fittedvalues attribute in a simple linear regression model provides the predicted y-values for each x-value in the dataset, calculated using the model’s intercept (𝛽0) and slope (𝛽1) from fitted_model.params or fitted_model.summary().tables[1]. The fitted line in a Simple Linear Regression model is chosen using the Ordinary Least Squares (OLS) method, which finds the line that best represents the relationship between the predictor x and outcome y in the observed data. OLS chooses the line that minimizes the sum of the squared differences between the observed y-values and the predicted y-values (the points on the line). These differences, known as residuals, represent the vertical distance between each observed data point and the line. By squaring these residuals, OLS ensures that all differences are positive. Larger residuals are penalized more heavily since larger errors have a disproportionately greater effect on the sum. This helps in finding a line that fits the data as closely as possible. Using squares instead of just absolute differences results in a unique, mathematically optimal solution for the line, balancing all residuals in a way that minimizes the overall error.

Explain why the first expression below can be interpreted as "the proportion of variation in (outcome) Y explained by the model (i.e. _fitted_model.fittedvalues_)"; and therefore, why _fitted_model.rsquared_ can be interpreted as a measure of the accuracy of the model; and, therefore what the two _np.corrcoef(...)[0,1]\*\*2_ expressions capture in the context of _Simple Linear Regression models_.

<details class="details-example"><summary style="color:blue"><u>Further Guidance</u></summary>

> _**R-squared** is the "the proportion of variation in (outcome) $Y$ explained by the model ($\hat y_i$)" and is defined as_
>
> _$R^2 = 1 - \frac{\sum_{i=1}^n(Y_i-\hat y)^2}{\sum_{i=1}^n(Y_i-\bar Y)^2}$_
>
> _The visuzation provided in the previous problem can be used to consider $(Y_i-\bar Y)^2$ as the squared distance of the $Y_i$ to their sample average $\bar Y$ as opposed to the squared **residuals** $(Y_i-\hat y)^2$ which is the squared distance of the $Y_i$ to their fitted (predicted) values $Y_i$._
>

The expressions provided all relate to R-squared, a key measure in regression analysis that quantifies the proportion of variation in the outcome variable \( Y \) explained by the model. 

- Expression 1, `1 - ((Y - fitted_model.fittedvalues)**2).sum() / ((Y - Y.mean())**2).sum()`, directly calculates R-squared by comparing the sum of squared residuals (the unexplained variation) to the total sum of squares (the total variation in Y). The ratio of these gives the proportion of variation not explained by the model, and subtracting it from 1 gives the proportion that is explained. 
- Expression 2, `fitted_model.rsquared`, directly provides the R-squared value, quantifying the accuracy of the model in explaining the variance in Y. 
- Expression 3, `np.corrcoef(Y, fitted_model.fittedvalues)[0,1]**2`, calculates the squared correlation between the observed and predicted values, which is equivalent to R-squared and reflects how well the model captures the relationship between Y and the fitted values. 
- Expression 4, `np.corrcoef(Y, x)[0,1]**2`, measures the squared correlation between the outcome variable Y and the predictor x, capturing the strength of their linear relationship, but it does not account for the model’s fit. While correlation and R-squared are related, the former simply measures the linear association between variables, whereas R-squared assesses the proportion of variation explained by the model.

### 8. Specify a *null hypothesis* of "no linear association (on average)" in terms of the relevant *parameter* of the *Simple Linear Regression* model, and use the code below to characterize the evidence in the data relative to the *null hypothesis* and interpret your subsequent beliefs regarding the Old Faithful Geyser dataset.<br>

<details class="details-example"><summary style="color:blue"><u>Further Guidance</u></summary>

> _Remember that **Hypothesis Testing** is not a "mathematical proof"_
>
> - _We do not prove $H_0$ false, we instead give evidence against the $H_0$: "We reject the null hypothesis with a p-value of XYZ, meaning we have ABC evidence against the null hypothesis"_

To test the null hypothesis of "no linear association (on average)" between waiting time and duration of the Old Faithful Geyser, we are testing if the slope of the regression line is equal to zero. This can be framed as:

Null Hypothesis (H0): The slope of the regression line is 0. There is no linear relationship between waiting time and eruption duration.
Alternative Hypothesis (HA): The slope of the regression line is not equal to 0. There is a linear relationship between waiting time and eruption duration.

Characterize the evidence:
- Step 1: Fit the model. 
  The code provided fits a linear regression model with duration as the dependent variable and waiting as the independent variable.
- Step 2: Review the summary. 
  The summary output from fitted_model.summary() will provide key statistics like the estimated coefficients, p-values, and confidence intervals.
- Step 3: Interpreting the Results.
  Coefficient for waiting represents the estimated change in eruption duration for each additional unit of waiting time. A significant coefficient would indicate that waiting time is a statistically significant predictor of eruption duration. P-value for waiting tells us whether the slope is significantly different from 0. If the p-value is smaller than the chosen significance level of 0.05, we reject the null hypothesis and conclude that there is a statistically significant linear relationship between waiting time and eruption duration. If the 95% confidence interval for the slope does not contain 0, it further supports rejecting the null hypothesis and suggests a significant linear relationship between the variables.

````python
import seaborn as sns
import statsmodels.formula.api as smf

# The "Classic" Old Faithful Geyser dataset
old_faithful = sns.load_dataset('geyser')

linear_for_specification = 'duration ~ waiting'
model = smf.ols(linear_for_specification, data=old_faithful)
fitted_model = model.fit()
fitted_model.summary()

# Extract key elements from the model summary
p_value = fitted_model.pvalues['waiting']
coef_waiting = fitted_model.params['waiting']
conf_int_waiting = fitted_model.conf_int().loc['waiting']
print(fitted_model.summary())
print('P-value:', p_value)
print('Coefficient:', round(coef_waiting,4))
print('Confidence Interval:', round(conf_int_waiting[0],4), ',', round(conf_int_waiting[1],4))

# Interpretation of the results
if p_value < 0.05:
    print("\nInterpretation:")
    print("The p-value for 'waiting' is less than 0.05, so we reject the null hypothesis.")
    print("There is a statistically significant linear relationship between the 'waiting' time and the 'duration' of eruptions.")
    print(f"The coefficient for 'waiting' is {coef_waiting:.4f}, meaning that for each additional minute of waiting time, the eruption duration increases by approximately {coef_waiting:.4f} minutes on average.")
    print(f"The 95% confidence interval for the coefficient of 'waiting' is [{conf_int_waiting[0]:.4f}, {conf_int_waiting[1]:.4f}], and since this interval does not contain zero, it supports the conclusion of a significant linear relationship.")
else:
    print("\nInterpretation:")
    print("The p-value for 'waiting' is greater than 0.05, so we fail to reject the null hypothesis.")
    print("There is no statistically significant linear relationship between the 'waiting' time and the 'duration' of eruptions.")
````

````python
import numpy as np
import pandas as pd
import seaborn as sns
import plotly.express as px
import statsmodels.formula.api as smf

long_wait_limit = 71
long_wait = old_faithful.waiting > long_wait_limit

print(smf.ols('duration ~ waiting', data=old_faithful[long_wait]).fit().summary().tables[1])

# Create a scatter plot with a linear regression trendline
fig = px.scatter(old_faithful[long_wait], x='waiting', y='duration', 
                 title="Old Faithful Geyser Eruptions for short wait times (>"+str(long_wait_limit)+")", 
                 trendline='ols')
fig.show(renderer='png')
````

````python
# Part 1
# Bootstrap sampling to collect the distribution of fitted slope coefficients
n_bootstrap = 1000  # Number of bootstrap samples
bootstrap_slopes = []

for _ in range(n_bootstrap):
    # Generate a bootstrap sample
    bootstrap_sample = old_faithful[long_wait].sample(frac=1, replace=True)
    
    # Fit the linear regression model to the bootstrap sample
    model = smf.ols('duration ~ waiting', data=bootstrap_sample).fit()
    
    # Collect the slope coefficient (beta1)
    bootstrap_slopes.append(model.params['waiting'])

# Visualize the bootstrap sampling distribution of the slope coefficients
sns.histplot(bootstrap_slopes, kde=True, color="blue", bins=30)
plt.title('Bootstrap Distribution of Slope Coefficients')
plt.xlabel('Slope Coefficient')
plt.ylabel('Frequency')
plt.show()
````

````python
# Part 2
# Set the true parameters for the simulation (Null hypothesis: no linear relationship)
beta_0 = 1.65
beta_1 = 0
sigma = 0.37
n_simulations = 1000  # Number of simulated datasets
simulated_slopes = []

# Get the 'waiting' values for the long wait times
x_values = old_faithful[long_wait]['waiting']

# Simulate Y values (duration) based on the null hypothesis parameters
for _ in range(n_simulations):
    # Simulate random noise with the specified standard deviation (sigma)
    noise = np.random.normal(0, sigma, size=len(x_values))
    
    # Simulate Y values using the model Y = beta_0 + beta_1 * X + noise
    y_simulated = beta_0 + beta_1 * x_values + noise
    
    # Fit the regression model to the simulated data
    model_sim = smf.ols('y_simulated ~ x_values', data=pd.DataFrame({'y_simulated': y_simulated, 'x_values': x_values})).fit()
    
    # Collect the slope coefficient (beta1)
    simulated_slopes.append(model_sim.params['x_values'])

# Visualize the distribution of the simulated slope coefficients
sns.histplot(simulated_slopes, kde=True, color="green", bins=30)
plt.title('Simulated Distribution of Slope Coefficients (Null Hypothesis)')
plt.xlabel('Slope Coefficient')
plt.ylabel('Frequency')
plt.show()
````

````python
# Part 3
# Compute the 95% confidence interval for the bootstrapped slope coefficients
lower_ci_boot, upper_ci_boot = np.percentile(bootstrap_slopes, [2.5, 97.5])
print(f"Simulated 95% Confidence Interval for Slope: ({lower_ci_sim:.4f}, {upper_ci_sim:.4f})")

# Compute the p-value from the simulated slopes (proportion of slopes >= 0)
p_value_simulated = np.mean(np.array(simulated_slopes) >= 0)
print(f"Simulated p-value: {p_value_simulated:.4f}")

# Fit the linear regression model to the actual long wait time data
actual_model = smf.ols('duration ~ waiting', data=old_faithful[long_wait]).fit()

# Get the p-value from the actual model's summary
actual_p_value = actual_model.pvalues['waiting']
print(f"Actual p-value from regression model: {actual_p_value:.4f}")

# Testing
if 0 >= lower_ci_boot and 0 <= upper_ci_boot:
    print('Zero is found in the bootstrapped 95% confidence interval')
    result1 = True
else:
    print('Zero is not in the bootstrapped 95% confidence interval')
    result1 = False
    
if p_value_simulated == actual_p_value:
    print("Actual p-value matches the p-value under null hypothesis simulation")
    result2 = True
else:
    print("Actual p-value does not match the p-value under null hypothesis simulation")
    result2 = False
    
if actual_p_value < 0.05:
    print("Actual p-value is less than significance level of 0.05")
    result3 = True
else:
    print("Actual p-value is greater than significance level of 0.05")
    result3 = False
    
if not(result1 and result2) and result3:
    print("Reject Null hypothesis")
else:
    print("Fail to reject Null Hypothesis")
````

### (A) Do a permuation test $\;H_0: \mu_{\text{short}}=\mu_{\text{long}} \; \text{ no difference in duration between short and long groups}$ by "shuffling" the labels
### (B) Create a 95% bootstrap confidence interval  by repeatedly bootstrapping within each group and applying *np.quantile(bootstrapped_mean_differences, [0.025, 0.975])* to the collection of differences between the sample means.

````python
import statsmodels.api as sm
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Load the Old Faithful dataset
old_faithful = sns.load_dataset('geyser')

# Separate the data into short and long waiting times (define short and long based on waiting time)
short_wait = old_faithful[old_faithful['waiting'] < old_faithful['waiting'].median()]
long_wait = old_faithful[old_faithful['waiting'] >= old_faithful['waiting'].median()]

# (A) Permutation Test for the Difference in Means
# Null Hypothesis: No difference in means between short and long waiting groups

# Calculate observed difference in means
observed_diff = short_wait['duration'].mean() - long_wait['duration'].mean()

# Combine the data and shuffle
combined = old_faithful['duration'].values
np.random.seed(42)  # For reproducibility
n_permutations = 10000
perm_diffs = []

for _ in range(n_permutations):
    np.random.shuffle(combined)  # Shuffle the combined dataset
    perm_short = combined[:len(short_wait)]  # First part for short wait
    perm_long = combined[len(short_wait):]  # Second part for long wait
    perm_diff = perm_short.mean() - perm_long.mean()
    perm_diffs.append(perm_diff)

# Calculate p-value from permutation distribution
perm_diffs = np.array(perm_diffs)
p_value = np.mean(np.abs(perm_diffs) >= np.abs(observed_diff))
print(f"Permutation Test p-value: {p_value}")

# Plot the permutation test distribution
plt.figure(figsize=(8, 6))
plt.hist(perm_diffs, bins=50, edgecolor='black', alpha=0.7)
plt.axvline(observed_diff, color='red', linestyle='--', label=f'Observed Diff: {observed_diff:.2f}')
plt.title('Permutation Test Distribution\n(Observed Diff vs Permuted Differences)')
plt.xlabel('Difference in Means')
plt.ylabel('Frequency')
plt.legend()
plt.show()

# (B) Bootstrap Confidence Interval for the Difference in Means
# Bootstrap sampling within each group
n_bootstrap = 10000
boot_diffs = []

for _ in range(n_bootstrap):
    boot_short = short_wait['duration'].sample(n=len(short_wait), replace=True)
    boot_long = long_wait['duration'].sample(n=len(long_wait), replace=True)
    boot_diff = boot_short.mean() - boot_long.mean()
    boot_diffs.append(boot_diff)

# Calculate the 95% bootstrap confidence interval
boot_diffs = np.array(boot_diffs)
lower, upper = np.quantile(boot_diffs, [0.025, 0.975])
print(f"Bootstrap Confidence Interval for Difference in Means: ({lower}, {upper})")

# Plot the bootstrap confidence interval
plt.figure(figsize=(8, 6))
plt.hist(boot_diffs, bins=50, edgecolor='black', alpha=0.7)
plt.axvline(lower, color='green', linestyle='--', label=f'Lower CI: {lower:.2f}')
plt.axvline(upper, color='green', linestyle='--', label=f'Upper CI: {upper:.2f}')
plt.axvline(observed_diff, color='red', linestyle='--', label=f'Observed Diff: {observed_diff:.2f}')
plt.title('Bootstrap Confidence Interval\n(Observed Diff vs Bootstrap CI)')
plt.xlabel('Difference in Means')
plt.ylabel('Frequency')
plt.legend()
plt.show()
````

(a) Explanation of Sampling Approaches

1. Permutation Test:
   - In a permutation test, we’re testing the null hypothesis H0: μ{short} = μ{long}, which assumes no difference in mean durations between the "short" and "long" groups.
   - Procedure: 
     - Combine the "short" and "long" groups into one pooled dataset.
     - Randomly shuffle the labels ("short" and "long") and reassign them to the observations.
     - Calculate the difference in mean duration between the two reshuffled groups.
     - Repeat the reshuffling many times to create a distribution of mean differences under the null hypothesis.
     - Determine the p-value by checking where the observed mean difference falls within this distribution.

2. Bootstrap Confidence Interval:
   - Bootstrapping generates a confidence interval for the difference in means by resampling with replacement within each group to estimate the sampling distribution of the difference in means.
   - Procedure:
     - Within each group ("short" and "long"), take repeated bootstrap samples (sampling with replacement).
     - For each bootstrap iteration, calculate the mean of each group and then find the difference in means.
     - After generating a collection of bootstrapped mean differences, use `np.quantile` to compute the 95% confidence interval by selecting the 2.5th and 97.5th percentiles of these differences.

(b) Comparison with Indicator Variable Model (from Question 11)

The indicator variable model involves creating a variable to distinguish between groups and fitting a model that estimates the effect of the group indicator on the outcome.

- Similarity:
  - Both methods aim to assess the difference in means between two groups and involve resampling to understand variability.
  - Each method provides insights into the effect of group classification (short vs. long wait times) on the outcome (duration).

- Differences:
  - Permutation Test and Bootstrap:
    - These methods do not assume a specific underlying model (they’re non-parametric).
    - They test or estimate the difference in means by resampling directly from the data, without needing an indicator variable in a model.
  - Indicator Variable Model:
    - This approach often involves fitting a regression model with a categorical indicator variable representing the groups.
    - It assumes a parametric relationship and can estimate the effect size (difference in means) while controlling for other variables if needed.

Both methods are useful for comparing groups but differ in the assumptions they require and how they interpret the group differences. The permutation test and bootstrap approaches are distribution-free, while the indicator variable model requires assumptions about the data’s distribution and structure.
